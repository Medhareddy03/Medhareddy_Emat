```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x="bmi", y="charges", hue="sex", alpha=0.5)
```

```python
sns.relplot(data=df, x="bmi", y="charges", col="sex", kind="scatter", hue="region", alpha=0.5)
```

```python
sns.relplot(data=df, x="bmi", y="charges", col="sex", kind="scatter", size="children", alpha=0.5)
```

```python
df.columns
```

```python
df["smoker_ind"] = df["smoker"].map({"yes": "Smoker", "no": "Non-Smoker"})
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df, columns=["smoker"])
```

```python
df.head()
```

```python
sns.relplot(data=df, x="bmi", y="charges", col="sex", kind="scatter", size="region", alpha=0.5)
```

```python
sns.relplot(data=df, x="bmi", y="charges", col="sex", kind="scatter", hue="smoker_ind", size="region", alpha=0.5)
```

```python
sns.pairplot(df, hue="smoker_ind")
```
