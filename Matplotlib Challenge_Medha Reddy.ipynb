```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
cces = pd.read_csv('cces_example_data.csv')  
```

```python
cces.head()
```

```python
cces.columns
```

```python
fig1, ax1 = plt.subplots(1, 2, figsize = (15, 5))  
ax1[0].set_title('Age Histogram') 
ax1[1].set_title('Political Identity') 
plt.show()
```

```python
cces.groupby('Evaluation')
```

```python
Evaluation_freq = cces.groupby('Evaluation')['caseid'].count()
Evaluation_freq
```

```python
fig2, ax2 = plt.subplots() 
ax2.bar(x = Evaluation_freq.index, height= Evaluation_freq)  
ax2.set_title('Evaluation Frequencies')
```

```python
fig3, ax3 = plt.subplots()   
ax3.scatter(x = cces['ConMediaUse'] , y = cces['ImmiPhobia'])
ax3.scatter(cces["ImmiPhobia"], cces["ConMediaUse"])
ax3.set_xlabel('ConMediaUse / ImmiPhobia')  
ax3.set_ylabel('ImmiPhobia / ConMediaUse') 
ax3.set_title('Scatter Plot of ConMediaUse & ImmiPhobia')
plt.show
```
